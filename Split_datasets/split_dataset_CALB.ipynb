```python
import numpy as np
import pickle
import pandas as pd
import numpy as np
import random
import os
import json
import json
from batteryml.data.battery_data import BatteryData
random.seed(2024)
np.random.seed(2024)
```

```python
# We manually split the dataset to avoid there is only one or two operating conditions in the testing set.
data_path = '/data/trf/python_works/Battery-LLM/dataset/CALB'
tmp_files = os.listdir(data_path)
life_labels = json.load(open('/data/trf/python_works/Battery-LLM/dataset/CALB_labels.json'))
files = [i for i in tmp_files if i in life_labels]

validation_set = ['CALB_0_B187.pkl', 'CALB_35_B222.pkl', 'CALB_25_T25-2.pkl', 'CALB_35_B247.pkl', 'CALB_45_B253.pkl']
testing_set = ['CALB_0_B182.pkl', 'CALB_25_T25-1.pkl', 'CALB_35_B174.pkl', 'CALB_35_B230.pkl', 'CALB_45_B255.pkl']
val_test_set = validation_set + testing_set
training_set = [i for i in files if i not in val_test_set]

for i in training_set:
    assert i not in validation_set
    assert i not in testing_set

for i in validation_set:
    assert i not in testing_set
    assert i not in training_set

for i in testing_set:
    assert i not in validation_set
    assert i not in training_set
    
print(f'training_set: {len(training_set)} {training_set}')
print(f'Val set: {len(validation_set)} {validation_set}')
print(f'Test set: {len(testing_set)} {testing_set}')
```
