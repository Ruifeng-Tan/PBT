# README
This notebook is used to generate domain ids fo each battery. The basic idea is retrieving the domain id from the prompts

```python
import sys
import os
import json
# Get the current notebook directory (process_scripts)
notebook_dir = os.getcwd()  # or os.path.dirname(os.path.abspath(''))
print(notebook_dir)
# Go up one level to BatteryMoE
parent_dir = os.path.dirname(notebook_dir)
print(parent_dir)

# Add BatteryMoE to Python's path
sys.path.append(parent_dir)
from Prompts.Mapping_helper import Mapping_helper
from data_provider.data_split_recorder import split_recorder
```

```python
Li_files = split_recorder.MIX_large_train_files + split_recorder.MIX_large_test_files + split_recorder.MIX_large_val_files
Na_files = split_recorder.NAion_2021_train_files + split_recorder.NAion_2021_test_files + split_recorder.NAion_2021_val_files
Zn_files = split_recorder.ZN_2024_train_files + split_recorder.ZN_2024_test_files + split_recorder.ZN_2024_val_files
CALB_files = split_recorder.CALB_train_files + split_recorder.CALB_test_files + split_recorder.CALB_val_files

all_files = Li_files + Na_files + Zn_files + CALB_files
```

```python
id = 0
existing_prompts = []
prompt2domainID = {}
for file in all_files:
    cell_name = file.split('.pkl')[0]
    helper = Mapping_helper(prompt_type='PROTOCOL', cell_name=cell_name)
    prompt = helper.do_mapping()
    if prompt not in existing_prompts:
        prompt2domainID[prompt] = id
        existing_prompts.append(prompt)
        id += 1
    else:
        continue

name2domainID = {}
for file in all_files:
    cell_name = file.split('.pkl')[0]
    helper = Mapping_helper(prompt_type='PROTOCOL', cell_name=cell_name)
    prompt = helper.do_mapping()
    name2domainID[cell_name] = prompt2domainID[prompt]
name2domainID
```

```python
save_path = '/data/trf/python_works/BatteryLife/dataset/'
with open(f'{save_path}name2domainID.json', 'w') as f:
    json.dump(name2domainID, f)
```
