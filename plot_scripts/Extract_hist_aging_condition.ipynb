# README
This notebook is used to extract the count of aging conditions. This prepares the data for histplot.

```python
import json
import pandas as pd
import os
import sys
import numpy as np
folder_a_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(folder_a_path)
from data_provider.data_split_recorder import split_recorder
```

```python
data_path = '../gate_data/name2agingConditionID.json'
tmp_data = json.load(open(data_path, 'r'))
data = {}
stanford_2_files = split_recorder.Stanford_formation_45_train_files_2021 + split_recorder.Stanford_formation_45_test_files_2021 + split_recorder.Stanford_formation_45_val_files_2021
stanford_1_files = split_recorder.Stanford_train_files + split_recorder.Stanford_val_files + split_recorder.Stanford_test_files
removed_files = [i for i in stanford_2_files if i not in stanford_1_files]
for key, value in tmp_data.items():
    if key in removed_files:
        continue
    data[key] = value
print(len(tmp_data))
print(len(data))
```

```python
aging_condition_ids = list(set([int(i) for i in data.values()]))
print(len(aging_condition_ids))
```

```python
df = pd.DataFrame()
agingconditionCount = {}
for name, agingConditionID in data.items():
    if agingConditionID not in agingconditionCount:
        agingconditionCount[agingConditionID] = 0
    agingconditionCount[agingConditionID] += 1

df['BatteryNumber'] = list(agingconditionCount.values())
df.to_excel('../figures/figures_data/agingConditionCount.xlsx', index=False)
```
