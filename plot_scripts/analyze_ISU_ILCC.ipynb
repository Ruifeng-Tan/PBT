Analyze the life distribution in the validation set of ISU_ILCC

```python
import json
import pandas as pd
import os
import sys
import matplotlib.pyplot as plt
from data_provider.data_split_recorder import split_recorder
import numpy as np
folder_a_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(folder_a_path)
```

```python
data_path = '/data/trf/python_works/BatteryLife/dataset/Life labels/ISU-ILCC_labels.json'
```

```python
label_data = json.load(open(data_path, 'r'))
label_data
```

```python
val_labels = {}
train_labels = {}
test_labels = {}
for key, value in label_data.items():
    if key in split_recorder.ISU_ILCC_val_files:
        val_labels[key] = value
    elif key in split_recorder.ISU_ILCC_train_files:
        train_labels[key] = value
    elif key in split_recorder.ISU_ILCC_test_files:
        test_labels[key] = value
val_labels
```

```python
val_labels_values = list(val_labels.values())
train_labels_values = list(train_labels.values())
test_labels_values = list(test_labels.values())
plt.figure(figsize=(10, 5))
plt.hist(val_labels_values)
plt.show()
plt.figure(figsize=(10, 5))
plt.hist(train_labels_values)
plt.show()
plt.figure(figsize=(10, 5))
plt.hist(test_labels_values)
plt.show()
```
