# README
This notebook is used to prepare Fig.2b data for originlab.

```python
import pandas as pd
import json
import os
import sys
import numpy as np
folder_a_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(folder_a_path)
```

```python
def get_mapes(data, cycle_number):
    mapes = np.array(data['mapes'])
    useable_cycle_numbers = np.array(data['Useable_cycle_number'])
    mapes  = np.mean(mapes[useable_cycle_numbers==cycle_number])
    return mapes

data_path = '../results'
dataset_name = 'CALB' # [CALB, NAion, ZN-coin]
CPMLP42_data = json.load(open(f'{data_path}/CPMLP_{dataset_name}42_42.json'))[f'{dataset_name}42']
CPMLP2021_data = json.load(open(f'{data_path}/CPMLP_{dataset_name}_2021.json'))[f'{dataset_name}']
CPMLP2024_data = json.load(open(f'{data_path}/CPMLP_{dataset_name}2024_2024.json'))[f'{dataset_name}2024']

CPTrans42_data = json.load(open(f'{data_path}/CPTransformer_{dataset_name}42_42.json'))[f'{dataset_name}42']
CPTrans2021_data = json.load(open(f'{data_path}/CPTransformer_{dataset_name}_2021.json'))[f'{dataset_name}']
CPTrans2024_data = json.load(open(f'{data_path}/CPTransformer_{dataset_name}2024_2024.json'))[f'{dataset_name}2024']

PBT42_data  = json.load(open(f'{data_path}/PBT_{dataset_name}42_42.json'))[f'{dataset_name}42']
PBT2021_data = json.load(open(f'{data_path}/PBT_{dataset_name}_2021.json'))[f'{dataset_name}']
PBT2024_data = json.load(open(f'{data_path}/PBT_{dataset_name}2024_2024.json'))[f'{dataset_name}2024']

df_data = {}
useable_cycle_numbers = list(set(PBT2021_data['Useable_cycle_number']))

PBT_means, PBT_stds = [], []
CPMLP_means, CPMLP_stds = [], []
CPTrans_means, CPTrans_stds = [], []
for cycle_number in useable_cycle_numbers:
    PBT_mean = np.mean([get_mapes(PBT42_data, cycle_number), get_mapes(PBT2021_data, cycle_number), get_mapes(PBT2024_data, cycle_number)])
    PBT_std = np.std([get_mapes(PBT42_data, cycle_number), get_mapes(PBT2021_data, cycle_number), get_mapes(PBT2024_data, cycle_number)])
    CPMLP_mean = np.mean([get_mapes(CPMLP42_data, cycle_number), get_mapes(CPMLP2021_data, cycle_number), get_mapes(CPMLP2024_data, cycle_number)])
    CPMLP_std = np.std([get_mapes(CPMLP42_data, cycle_number), get_mapes(CPMLP2021_data, cycle_number), get_mapes(CPMLP2024_data, cycle_number)])
    CPTrans_mean = np.mean([get_mapes(CPTrans42_data, cycle_number), get_mapes(CPTrans2021_data, cycle_number), get_mapes(CPTrans2024_data, cycle_number)])
    CPTrans_std = np.std([get_mapes(CPTrans42_data, cycle_number), get_mapes(CPTrans2021_data, cycle_number), get_mapes(CPTrans2024_data, cycle_number)])
    PBT_means.append(PBT_mean)
    PBT_stds.append(PBT_std)
    CPMLP_means.append(CPMLP_mean)
    CPMLP_stds.append(CPMLP_std)
    CPTrans_means.append(CPTrans_mean)
    CPTrans_stds.append(CPTrans_std)

df_data['cycle_number'] = useable_cycle_numbers
df_data['PBT'] = PBT_means
df_data['PBT_std'] = PBT_stds
df_data['CPMLP'] = CPMLP_means
df_data['CPMLP_std'] = CPMLP_stds
df_data['CPTrans'] = CPTrans_means
df_data['CPTrans_std'] = CPTrans_stds

df = pd.DataFrame(df_data)
df.to_excel(f'/data/trf/python_works/BatteryMoE/results/fig5_{dataset_name}_data.xlsx', index=False)
df.head(100)
```
