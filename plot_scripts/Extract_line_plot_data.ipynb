This notebook is used to prepare Fig.2b data for originlab.

```python
import pandas as pd
import json
import os
import sys
import numpy as np
folder_a_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(folder_a_path)
```

```python
CPMLP42_data = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_CPMLP_MIX_large_MIX_large_42.json'))
CPMLP2021_data = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_CPMLP_MIX_large_MIX_large_2021.json'))
CPMLP2024_data = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_CPMLP_MIX_large_MIX_large_2024.json'))

CPTrans42_data = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_CPTransformer_MIX_large_MIX_large_42.json'))
CPTrans2021_data = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_CPTransformer_MIX_large_MIX_large_2021.json'))
CPTrans2024_data = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_CPTransformer_MIX_large_MIX_large_2024.json'))

PBT42_data  = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_PBT_MIX_large_MIX_large_42.json'))
PBT2021_data = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_PBT_MIX_large_MIX_large_2021.json'))
PBT2024_data = json.load(open('/data/trf/python_works/BatteryMoE/output_path/number_MAPE_PBT_MIX_large_MIX_large_2024.json'))

df_data = {}
useable_cycle_numbers = list(PBT2021_data.keys())
PBT_means, PBT_stds = [], []
CPMLP_means, CPMLP_stds = [], []
CPTrans_means, CPTrans_stds = [], []
for cycle_number in useable_cycle_numbers:
    PBT_mean = np.mean([PBT42_data[cycle_number], PBT2021_data[cycle_number], PBT2024_data[cycle_number]])
    PBT_std = np.std([PBT42_data[cycle_number], PBT2021_data[cycle_number], PBT2024_data[cycle_number]])
    CPMLP_mean = np.mean([CPMLP42_data[cycle_number], CPMLP2021_data[cycle_number], CPMLP2024_data[cycle_number]])
    CPMLP_std = np.std([CPMLP42_data[cycle_number], CPMLP2021_data[cycle_number], CPMLP2024_data[cycle_number]])
    CPTrans_mean = np.mean([CPTrans42_data[cycle_number], CPTrans2021_data[cycle_number], CPTrans2024_data[cycle_number]])
    CPTrans_std = np.std([CPTrans42_data[cycle_number], CPTrans2021_data[cycle_number], CPTrans2024_data[cycle_number]])

    PBT_means.append(PBT_mean)
    PBT_stds.append(PBT_std)
    CPMLP_means.append(CPMLP_mean)
    CPMLP_stds.append(CPMLP_std)
    CPTrans_means.append(CPTrans_mean)
    CPTrans_stds.append(CPTrans_std)

df_data['cycle_number'] = useable_cycle_numbers
df_data['PBT'] = PBT_means
df_data['PBT_std'] = PBT_stds
df_data['CPMLP'] = CPMLP_means
df_data['CPMLP_std'] = CPMLP_stds
df_data['CPTrans'] = CPTrans_means
df_data['CPTrans_std'] = CPTrans_stds

df = pd.DataFrame(df_data)
df.to_csv('/data/trf/python_works/BatteryMoE/results/fig2b_data.csv', index=False)
```
