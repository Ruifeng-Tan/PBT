```python
import os
import sys
folder_a_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(folder_a_path)
import matplotlib.pyplot as plt
import numpy as np
from data_provider.data_split_recorder import split_recorder
import json
import pandas as pd
```

```python
dataset_names = ['HNEI', 'MATR', 'MICH', 'XJTU', 'Stanford', 'RWTH', 'MICH-EXP', 'Tongji', 
                 'HUST', 'SNL', 'ISU-ILCC', 'CALCE', 'UL-PUR', 'CALB', 'NA-ion', 'ZN-coin']

data_path = '/data/trf/python_works/BatteryLife/dataset/Life labels/'

for i, dataset_name in enumerate(dataset_names):

    # --- 处理 MICH / MICH-EXP 特殊情况 ---
    if dataset_name in ['MICH', 'MICH-EXP']:
        dataset_name_key = 'total_MICH'
        if dataset_name == 'MICH-EXP':
            cell_names = split_recorder.MICH_EXP_train_files + \
                         split_recorder.MICH_EXP_test_files + \
                         split_recorder.MICH_EXP_val_files
        else:
            cell_names = split_recorder.MICH_train_files + \
                         split_recorder.MICH_test_files + \
                         split_recorder.MICH_val_files
    else:
        dataset_name_key = dataset_name

    life_label = json.load(open(f'{data_path}{dataset_name_key}_labels.json', 'r'))

    if dataset_name not in ['MICH', 'MICH-EXP']:
        life_labels = list(life_label.values())
    else:
        life_labels = [life_label[c] for c in cell_names if c in life_label]
    
    if dataset_name == 'HNEI':
        training_cell_names = split_recorder.HNEI_train_files
    elif dataset_name == 'MATR':
        training_cell_names = split_recorder.MATR_train_files  
    elif dataset_name == 'MICH':
        training_cell_names = split_recorder.MICH_train_files
    elif dataset_name == 'MICH-EXP':
        training_cell_names = split_recorder.MICH_EXP_train_files
    elif dataset_name == 'XJTU':
        training_cell_names = split_recorder.XJTU_train_files
    elif dataset_name == 'Stanford':
        training_cell_names = split_recorder.Stanford_train_files
    elif dataset_name == 'RWTH':
        training_cell_names = split_recorder.RWTH_train_files
    elif dataset_name == 'Tongji':
        training_cell_names = split_recorder.Tongji_train_files
        training_cell_names = [c.replace('--', '-#') for c in training_cell_names]
    elif dataset_name == 'HUST':
        training_cell_names = split_recorder.HUST_train_files
    elif dataset_name == 'SNL':
        training_cell_names = split_recorder.SNL_train_files
    elif dataset_name == 'ISU-ILCC':
        training_cell_names = split_recorder.ISU_ILCC_train_files
    elif dataset_name == 'CALCE':
        training_cell_names = split_recorder.CALCE_train_files
    elif dataset_name == 'UL-PUR':
        training_cell_names = split_recorder.UL_PUR_train_files
    elif dataset_name == 'CALB':
        training_cell_names = split_recorder.CALB_train_files
    elif dataset_name == 'NA-ion':
        training_cell_names = split_recorder.NAion_2021_train_files
    elif dataset_name == 'ZN-coin':
        training_cell_names = split_recorder.ZNcoin_train_files
    
    cell_have_labels = [c for c in training_cell_names if c in life_label]
    print(f'{dataset_name} has {len(cell_have_labels)} batteries with labels')
```
