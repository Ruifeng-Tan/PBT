# README
This notebook is used to plot the model performance versus seen cycle number.

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(4,3))
BiGRU_MAPE = [0.201148765, 0.192504848, 0.19183801, 0.202307638, 0.201665543]
iTransformer = [0.232451429, 0.189159401, 0.179470253, 0.185392971, 0.185974077]
BatteryLifeLLM = [0.1747, 0.1615, 0.159183, 0.1609923, 0.161335]
Xs = [1, 2, 3, 4, 5]
names = ['1-10', '25-35', '50-60', '75-85', '90-100']
plt.plot(Xs, BiGRU_MAPE, label='BiGRU', marker='o', markerfacecolor='white')
plt.plot(Xs, iTransformer, label='iTransformer', marker='o', markerfacecolor='white')
plt.plot(Xs, BatteryLifeLLM, label='BatteryLifeLLM', marker='o', markerfacecolor='white')
plt.xticks(Xs, names, rotation=45)
plt.ylabel('MAPE', fontsize=10)
plt.xlabel('Seen cycle number', fontsize=10)
plt.legend()
fig.tight_layout()
plt.savefig('figures/model_performance.jpg', dpi=600)
plt.show()
```
